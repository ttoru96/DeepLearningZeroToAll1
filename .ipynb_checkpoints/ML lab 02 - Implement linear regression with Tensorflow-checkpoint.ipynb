```python
import tensorflow as tf

# tensorflow.Variable: 
# <=> tensorflow가 train하는 variable이다.
# <=> trainable한 variable이다.

# 1. Build graph using TF operations
# X and Y data
x_train = [1, 2, 3]
y_train = [1, 2, 3]

# Hyperparameters W and b
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis function: H(x) = Wx + b
# hypothesis = W*x_train + b
hypothesis = x_train*W + b

# cost function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost) # train is a node

# 2. Run/update graph and get results
# Launch the graph in a session.
sess = tf.Session()
# Initializes global variables in the graph
sess.run(tf.global_variables_initializer())

# Fit the line
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
        # cost가 아닌 sess.run(cost)를 해야만 cost의 값이 evaluate 되는 것이다
```
